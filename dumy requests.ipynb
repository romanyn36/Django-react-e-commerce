# Make POST request with fake data

```python

from requests_toolbelt.multipart.encoder import MultipartEncoder

fake = Faker()

def create_dummy_products(n):
    headers = {
        'Authorization': f'Bearer {TOKEN}',
    }

    for _ in range(n):
        data = {
            'name': fake.word(),
            'price': round(random.uniform(10.0, 100.0), 2),
            'brand': fake.company(),
            'countInStock': random.randint(0, 100),
            'category': fake.word(),
            'description': fake.text(),
        }

        multipart_data = MultipartEncoder(
            fields={
                'name': data['name'],
                'price': str(data['price']),  # Convert to string
                'brand': data['brand'],
                'countInStock': str(data['countInStock']),  # Convert to string
                'category': data['category'],
                'description': data['description'],
                # 'image': ('', '', 'application/octet-stream')  # Dummy file part
            }
        )

        headers['Content-Type'] = multipart_data.content_type

        response = requests.post(URL, headers=headers, data=multipart_data)

        if response.status_code == 201:
            print(f'Successfully created product: {response.json()}')
        else:
            print(f'Failed to create product: {response.status_code}, {response.text}')

# Number of dummy products to create
create_dummy_products(10)
```
